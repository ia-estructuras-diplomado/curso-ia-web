## 3. Transformar datos (list comprehension)

Sintaxis compacta: `[expr for x in lista if condición]`.


### 📘 Preguntas teóricas — 3 — Comprehension

1. ¿Qué partes tiene `[precio * 0.9 for precio in precios if precio > 100]`?
2. ¿Cuándo conviene un `for` explícito en lugar de comprehension en un pipeline de ML?


#### ✍️ Tus respuestas — 3 — Comprehension

1.

````python
precios = [100, 250, 45, 800, 120]
print('Originales:', precios)
````

````python
### TU TAREA AQUÍ ###
UMBRAL_DESCUENTO = 100
precios_con_descuento = [
    precio * 0.9 if precio > UMBRAL_DESCUENTO else precio
    for precio in precios
]
print('Tras regla:', precios_con_descuento)
````

````python
r = verificar_precios(precios_con_descuento, UMBRAL_DESCUENTO)
resumen_seccion('3 — Comprehension', r)
n_desc = sum(1 for p in precios if p > UMBRAL_DESCUENTO)
print(f"ℹ️ {n_desc} producto(s) con descuento.")
````

### 📘 Preguntas teóricas — 4 — Funciones

````python
def predecir_riesgo(edad: int, ingresos: float) -> str:
    """Evalúa riesgo: Bajo / Medio / Alto según reglas fijas."""
    if ingresos > 50000 and edad > 25:
        return "Riesgo Bajo"
    if ingresos < 20000:
        return "Riesgo Alto"
    return "Riesgo Medio"
````

````python
### TU TAREA AQUÍ ###
INGRESOS_PARA_PROBAR = 60_000  # 15_000 → Alto | 30_000 → Medio
resultado_agente = predecir_riesgo(perfil_usuario["edad"], INGRESOS_PARA_PROBAR)
print('Tool →', resultado_agente)
````

````python
_mapa = {60_000: 'Riesgo Bajo', 15_000: 'Riesgo Alto', 30_000: 'Riesgo Medio'}
esperado = _mapa.get(INGRESOS_PARA_PROBAR)
if esperado:
    r = verificar_riesgo(resultado_agente, INGRESOS_PARA_PROBAR, esperado)
else:
    print("⚠️ Usa 60000, 15000 o 30000 para autoevaluación.")
    r = [True]
resumen_seccion('4 — Funciones', r)
````
