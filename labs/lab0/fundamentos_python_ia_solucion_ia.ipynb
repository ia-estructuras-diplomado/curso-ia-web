## 0. Sintaxis esencial de Python

Antes de ML o agentes, necesitas leer y escribir código Python básico.

| Elemento | Ejemplo | Nota para IA |
|----------|---------|--------------|
| Comentario | `# esto no se ejecuta` | Documentar pasos del pipeline |

## 4. Funciones = herramientas para agentes

Sintaxis: `def nombre(param: tipo) -> tipo:` + docstring + `return`.


#### ✅ Respuestas sugeridas — 4 — Funciones

1. String triple comilla justo bajo `def`; documenta uso para humanos y agentes.
2. Type hints: contrato, autocompletado, linters (mypy); buena práctica en código de producción.
3. El modelo elige qué tool llamar leyendo nombres y descripciones.


````python
def predecir_riesgo(edad: int, ingresos: float) -> str:
    """Evalúa riesgo: Bajo / Medio / Alto según reglas fijas."""
    if ingresos > 50000 and edad > 25:
        return "Riesgo Bajo"
    if ingresos < 20000:
        return "Riesgo Alto"
    return "Riesgo Medio"
````

### 🤖 Sección 4 — Funciones

**Objetivo:** Llamar predecir_riesgo como tool de agente.

**Lista de tareas**

**Prompt sugerido** (copiar al asistente):

```text

resultado_agente = predecir_riesgo(perfil_usuario["edad"], INGRESOS_PARA_PROBAR)

```

_Puedes usar GitHub Copilot, Gemini, ChatGPT o Cursor. Si falla la autoevaluación, pide a la IA que corrija usando el mensaje ❌._


````python
### Referencia docente (pasa ✅) ###
INGRESOS_PARA_PROBAR = 60_000
resultado_agente = predecir_riesgo(perfil_usuario["edad"], INGRESOS_PARA_PROBAR)
print('Tool →', resultado_agente)
````
